# Exploring weather trends

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loding the dataset
city_df=pd.read_csv('city.csv')
global_df=pd.read_csv('global_data.csv')
```

```python
# Checking for null values
city_df.isnull().sum()
```

```python
# Checking for null values
global_df.isnull().sum()
```

```python
# Filling nan values with mean values
mean=city_df['avg_temp'].mean()
city_df['avg_temp'].fillna(mean,inplace=True)
```

```python
# Doing moving average for city and global data
global_moving_avg=global_df['avg_temp'].rolling(5).mean()
city_moving_avg=city_df['avg_temp'].rolling(5).mean()
```

```python
# Plotting the data
plt.plot(global_df['year'],global_moving_avg,label='global');
plt.plot(city_df['year'],city_moving_avg,label='local');
plt.xlabel("Years");
plt.ylabel("Temperature (°C)");
plt.legend();
plt.title("Kanpur city(India) temp w.r.t global temp ");
```
